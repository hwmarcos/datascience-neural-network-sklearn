```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
print(cancer.feature_names)
print(cancer.target_names)
print(len(cancer))
print(len(cancer.data))
print(cancer.keys())
```

```python
print(cancer.DESCR)
```

```python
#pegando as entradas x e as saídas y
x, y = cancer.data, cancer.target
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
#75% do dataset para TREINO e 25% para TESTE
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y)
print(len(x_treino))
print(len(x_teste))
```

```python
#normalizando os dados
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(x_treino)
x_treino = scaler.transform(x_treino)
x_teste = scaler.transform(x_teste)
```

```python
#treinando o modelo
from sklearn.neural_network import MLPClassifier
```

```python
#hidden_layer_sizes: 3 camadas escondidas com 30 neuroios cada
#max_iter: máximo de épocas
mlp = MLPClassifier(hidden_layer_sizes=(30,30,30), max_iter=1000)
mlp.fit(x_treino, y_treino)
```

```python
prediction = mlp.predict(x_teste)
prediction
```

```python
score = mlp.score(x_teste, y_teste) * 100
score
```

```python
#avaliando o modelo
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_teste, prediction))
```

```python
print(classification_report(y_teste, prediction))
```
